GRADIENT DESCENT ALGORITHM

```python
def log_loss(y_true, y_prediction):
    epsilon = 1e-15
    y_prediction_new = [max(i, epsilon) for i in y_prediction]
    y_prediction_new = [min(i, 1-epsilon) for i in y_prediction_new]
    y_prediction_new = np.array(y_prediction_new)
    return -np.mean(y_true * np.log(y_prediction_new) + (1-y_true) * np.log(1-y_prediction_new))
```

```python
def sig_num(x):
    return 1 / (1+np.exp(-x))
sig_num(np.array([12,0,1]))
```

```python
def gradient(age, affordability, y_true, epochs, loss_threashold):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)

    for i in range(epochs):
        sum =  w1 * age + w2 * affordability + bias
        y_pred = sig_num(sum)
        loss = log_loss(y_true, y_pred)
        wd = (1/n)*np.dot(np.transpose(age),  (y_pred - y_true))
        wd2 = (1/n)*np.dot(np.transpose(affordability),  (y_pred - y_true))
        bias_d = np.mean(y_pred - y_true)
        w1 = w1 - rate * wd
        w2 = w2 - rate * wd2

        bias = bias - rate * bias_d

        print(f"Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}")
        if loss <= loss_threashold:
            break
    return w1, w2, bias
```

```python
class myNN:
    def __init__(self):
        self.w1=  1
        self.w2 = 1
        self.bias = 0
    def fit(self, x, y, epochs, loss_threashold):
        self.w1, self.w2, self.bias = self.gradient(x['age'], x['affordability'], y, epochs, loss_threashold)
    def predict(self, x_test):
        w_s = self.w1 * x_test['age'] + self.w2 * x_test['affordability'] + self.bias
        return sig_num(w_s)
    def gradient(self, age, affordability, y_true, epochs, loss_threashold):
        w1 = w2 = 1
        bias = 0
        rate = 0.5
        n = len(age)

        for i in range(epochs):
            sum = w1 * age + w2 * affordability + bias
            y_pred = sig_num(sum)
            loss = log_loss(y_true, y_pred)
            wd = (1/n) * np.dot(np.transpose(age), (y_pred - y_true))
            wd2 = (1/n) * np.dot(np.transpose(affordability), (y_pred - y_true))
            bias_d = np.mean(y_pred - y_true)
            w1 = w1 - rate * wd
            w2 = w2 - rate * wd2
            bias = bias - rate * bias_d
            if i % 50 == 0:
                print(f"Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}")
            if loss <= loss_threashold:
                 print(f"Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}")
                 break
                

        return w1, w2, bias
```
